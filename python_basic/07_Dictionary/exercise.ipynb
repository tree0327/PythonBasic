```python
# 1.shallow copy (얕은 복사)
a = {'a': 1, 'b': 2, 'c': 3}
b = a # 사전 레퍼런스의 복사 (공유됨)
print(a)
print(b)

a['a'] = 100
print(a)
print(b)
```

```python
import copy

original = {
    'name' : 'capybara',
    'score' : [ 100, 90, 80 ]
}

deep = copy.deepcopy(original)

original['score'][0] = 999

print("original : ", original)
print("deep : ", deep)
```

```python
dic1 = {
    'a' : 10,
    'b' : 20
    }

dic2 = {
    'c' : 100,
    'd' : 300
    }

dic1.update(dic2)

print(dic1)
print(dic2)
```

```python
# dict의 정렬 : sorted
numbers = {
    'first': 1,
    'second': 2,
    'third': 3,
    'Fourth': 4
    }

numbers['hundred'] = 100
numbers['fifth'] = 5

print(numbers)
print(list(numbers))

print(sorted(numbers))
```
