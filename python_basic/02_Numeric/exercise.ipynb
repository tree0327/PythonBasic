### 정수값 = **int**

```python
num1 = 1
num2 = 2
print(type(num1))
print(type(num2))
```

```python
# 정수 진법별 접두사
# 2진수 (0, 1) : 0b
# 8진수 (0 ~ 7) : 0o
# 16진수 (0 ~ 15) : 0x

a = 100
print(a)

b = 0o23 # 8진수
print(b)

c = 0xff # 16진수
print(c)
```

### 실수값 = **float**

```python
num3 = 3.14
num4 = 1.0
print(type(num3))
print(type(num4))
```

```python
# 지수표현식
# 4.24E10 => 4.24 * 10*10, 4.24e-10 => 4.24 * 10**-10

a = 4.24e3
b = 4.24e-2
print(a)
print(b)
```

### 복소수형 = complex

```python
num5 = 3 + 4j
print(type(num5))
```

```python
num1 = 10
num2 = 3

print(num1 + num2)
print(num1 - num2)
print(num1 / num2)
print(num1 * num2)
```

```python
# 자동형변환 (실수와 정수의 계산에서는 정수를 실수로 자동변환함)
num3 = 10
num4 = 0.1
print(num3 + num4)
print(num3 - num4)
print(num3 / num4)
print(num3 * num4)
```

```python
num1 = 11
num2 = 7
print(num1 % num2)
```

```python
num1 = 11
num2 = 7
print(num1 // num2)
```

```python
base = 9
exponent = 2
result = base ** exponent

print(result)
```
