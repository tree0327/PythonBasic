### 정수값 = **int**

```python
num1 = 1
num2 = 2
print(type(num1))
print(type(num2))
```

```python
# 정수 진법별 접두사
# 2진수 (0, 1) : 0b
# 8진수 (0 ~ 7) : 0o
# 16진수 (0 ~ 15) : 0x

a = 100
print(a)

b = 0o23 # 8진수
print(b)

c = 0xff # 16진수
print(c)
```

### 실수값 = **float**

```python
num3 = 3.14
num4 = 1.0
print(type(num3))
print(type(num4))
```

```python
# 지수표현식
# 4.24E10 => 4.24 * 10*10, 4.24e-10 => 4.24 * 10**-10

a = 4.24e3
b = 4.24e-2
print(a)
print(b)
```
