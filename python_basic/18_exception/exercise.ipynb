```python
class MyError(Exception):
    pass

def say_nick(nick):
    if nick == '돼지':
        raise MyError()
    print(nick)

try:
    say_nick("곰")
    say_nick("돼지")

except MyError:
    print('허용하지 않는 별명입니다.')
```

```python
try:
    4 / 0 # ZeroDivisionError
  # 4 / 1 오류가 아님
except ZeroDivisionError:
    print('예외 발생')
```

```python
try:
    a = [ 1, 2 ]
    print(a[3]) # 첫번째 에러
    4/0 # 두번째 에러 (첫번째에서 에러가 났기때문에 굳이 실행되지 않음)
    
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except IndexError:
    print("인덱싱 할 수 없습니다.")

# 실행 결과: 인덱싱 할 수 없습니다.
# 에러가 발생한 줄에서 종료됨 (interpreter)
```

```python
try:
    4 / 0
except ZeroDivisionError as e:
    print(e)

# 실행 결과: division by zero
```

```python
try:
    print('파일 열기')
    f = open('foo.txt', 'z')
    # 동작할 코드 내용 작성    

finally:
    print('파일 닫기')
    f.close()    # try 블럭 수행 중간에 오류가 발생하더라도 마지막에 무조건 실행된다.
```

```python
try:
    age=int(input('나이를 입력하세요: '))

except:
    print('입력이 정확하지 않습니다.')

else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
```
