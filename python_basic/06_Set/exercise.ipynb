```python
#remove() = 요소삭제, 값이 존재하지 않으면 에러
#discard() = 요소삭제, 값이 존재하지 않아도 에러 x

ohgiraffers = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}
# ohgiraffers.remove('Elephant')    # Error 발생
ohgiraffers.remove('Pig')
print(ohgiraffers)    #  {'Bear', 'Squirrel', 'Gorilla'}
ohgiraffers.discard('Pig')
print(ohgiraffers)
```

```python
ohgiraffers.pop()
print(ohgiraffers) # 순서가 없기에 랜덤으로 삭제됨

# 전부 다 지우려면 clear()
```

```python
javaTeam = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
pythonTeam = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

# union 합집합
ohgiraffers = javaTeam.union(pythonTeam)
print(ohgiraffers)

# | -> or과 같은 개념 // 두 집합 중에 하나라도 있으면 포함
new_ohgiraffers = javaTeam | pythonTeam
print(new_ohgiraffers)
```

```python
# intersection 교집합
print(javaTeam.intersection(pythonTeam))

# & -> and와 같은 개념 // 두 집합 모두에 있으면 포함
print(javaTeam & pythonTeam)
```

```python
# difference 차집합
print(javaTeam.difference(pythonTeam))

# - -> 보이는 그대로
print(javaTeam - pythonTeam)
```

```python
# symmetric_difference 공통차집합 // 공통이지 않은것들의 전부

print(javaTeam.symmetric_difference(pythonTeam))
print(javaTeam ^ pythonTeam)
```
