### 문제 2

다음 조건에 따라 **비교 연산 결과를 출력하는 코드**를 작성하세요.

1. 문자열과 숫자를 이용하여 다양한 비교 연산을 수행해보세요.
2. 각 비교식의 결과를 `print()` 함수를 사용하여 출력하세요.
3. 출력해야 하는 비교식은 다음과 같습니다.
    - `'hello'` 와 `'Hello'` 의 동일 여부
    - `100` 과 `100` 의 서로 다름 여부
    - `10` 과 `9` 의 대소 비교
    - `'a'` 와 `'A'` 의 대소 비교(문자 비교)

```python
print('hello' == 'Hello')
print(100 != 100)
print(10 > 9)
print('a' > 'A')
```

### 문제 3

다음 조건에 맞게 변수를 선언하고, 주어진 논리 조건식의 결과를 출력하세요.

1. `name` 에 문자열 `'카피바라'`를 저장하세요.
2. `age` 에 정수 `20`을 저장하세요.
3. `gender` 에 문자열 `'male'`을 저장하세요.
4. 아래 조건을 각각 검사하여 `print()`로 결과를 출력하세요.
    - 이름이 `'카피바라'`이면 `True`
    - 나이가 **19세 이상이면 성인**으로 판단하고, 성인이면 `True`
    - 성인이면서 남성이면 `True`
    - 이름이 `'호랑이'`이고, 성인이며, **남성**이면 `True`
    - 이름이 `'카피바라'`이거나 성인이면 `True`
    - 이름이 `'카피바라'`가 아니라면 `True` (`not` 사용)

```python
name = '카피바라'
age = 20
gender = 'male'

adult = age >= 19

print(name == "카피바라") # 1
print(adult) # 2
print(adult and gender == 'male') # 3
print(name == '호랑이' and adult and gender == 'male') # 4
print(name == '카피바라' or adult) # 5
print(not name == '카피바라') # 6
```
