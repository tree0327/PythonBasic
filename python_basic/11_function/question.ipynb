## **문제 4: 점심 메뉴 추천**

**"점심시간, 다람쥐는 팀원들에게 점심 메뉴를 추천해준다."**

1. `recommend_menu(*menus)`라는 함수를 만들어 여러 개의 점심 메뉴를 매개변수로 전달받으세요.
2. 랜덤하게 하나의 메뉴를 선택하고 반환하세요.
    - 출력 예시: **"오늘의 점심 메뉴는 [선택된 메뉴]입니다!"**
    - 힌트: `random.choice()`를 사용하면 랜덤 선택이 가능합니다.

```python
import random

def recommend_menu(*menus):
    today_menu = random.choice(menus)

    print(f"오늘의 점심 메뉴는 {today_menu}입니다.")

recommend_menu('삼겹살', '피자', '치킨', '탕수육', '짜장면')
```

## **문제 6: 업무 성과 분석하기**

**"다람쥐는 오늘의 업무 성과를 보고해야 한다."**

1. `work_report(tasks_done)`이라는 함수를 만들어 오늘 처리한 업무 수를 매개변수로 받으세요.
2. 함수는 다음 두 값을 반환합니다.
    - `tasks_done`: 처리한 업무 수
    - `status`: 업무 성과 평가 (10개 이상: "우수", 5~9개: "보통", 4개 이하: "노력 필요")
3. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"오늘 처리한 업무는 [업무 수]개입니다. 성과 평가: [평가]"**

```python
# match

def work_report(tasks_done):
    match tasks_done:
        case i if i >= 10:
            status = "우수"

        case i if i >= 5:
            status = "보통"

        case _:
            status = "노력 필요"

    return f"오늘 처리한 업무는 {tasks_done}개입니다. 성과 평가: {status}"

work_report(10)
```

```python
# if

def work_report(tasks_done):
    if tasks_done >= 10:
        status = "우수"
    elif tasks_done >= 5:
        status = "보통"
    elif tasks_done <= 4:
        status = "노력 필요"

    print(f"오늘 처리한 업무는 {tasks_done}개입니다. 성과 평가: {status}")

work_report(5)
```

2. 총 월급을 계산하고, 세금은 총 월급의 10%로 공제하세요.
3. 함수는 다음 두 값을 반환합니다.
    - `net_salary`: 실수령액 (총 월급 - 세금)２
    - `tax`: 세금 공제액
4. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"세전 월급: [총 월급]원, 세금: [세금]원, 실수령액: [실수령액]원"**

```python
# 1. calculate 함수 만들거다 (입력 2개)
    # 2. 이거 두개 합쳐서 총월급
    # 3. 세금은 총월급의 10%
    # 4. 실수령액은 총월급 - 세금
    # 5. 출력값은 총월급, 세금, 실수령액
    
# 함수를 호출 (입력 2개 - 기본월급, 보너스) 해서 나온 값 저장

# 출력 (중간에 총월급, 세금, 실수령액 3개 f 스트링으로 넣으면 될것같다.)
```

```python
def calculate_salary(base_salary, bonus):
    total = base_salary + bonus
    tax = total * 0.1
    net_salary = total - tax

    return total, tax, net_salary

result = calculate_salary(10_000_000, 2_500_000)

print(f"세전 월급: {result[0]}원, 세금: {result[1]}원, 실수령액: {result[2]}원")
```
