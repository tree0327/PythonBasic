### **문제 2: `.csv` 파일 생성 및 읽기**

다람쥐는 이직 준비를 위해 지원했던 회사들의 정보를 저장하려고 합니다.

1. `applications.csv` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```
    회사명,직무,결과
    회사A,데이터 분석가,합격
    회사B,소프트웨어 엔지니어,불합격
    회사C,제품 관리자,진행 중
    ```
    
2. 파일을 닫고 다시 열어 모든 데이터를 출력하세요.

````python
import csv

with open("./quest/applications.csv", 'w') as f:
    f.write("회사명, 직무, 결과\n회사A,데이터 분석가,합격\n회사B,소프트웨어 엔지니어,불합격\n회사C,제품 관리자,진행중")

with open("./quest/applications.csv", 'r') as f_open:
    # print(f_open.read())
    reader = csv.reader(f_open)
    # print(reader)
    for i in reader:
        print(",".join(i))
````

````python
# csv 형태로 저장하기
import csv

data = [
    ["회사명", "직무", "결과"],
    ["회사A","데이터 분석가","합격"],
    ["회사B","소프트웨어 엔지니어","불합격"],
    ["회사C","제품 관리자","진행중"]
]

with open("applications.csv", 'w', newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(data)

with open("applications.csv", 'r') as f_open:
    reader = csv.reader(f_open)
    for i in reader:
        print(",".join(i))
````

````python
# 딕셔너리 형태로 저장하기

import csv

data = [
    {"회사명": "회사A", "직무": "데이터 분석가", "결과": "합격"},
    {"회사명": "회사B", "직무": "소프트웨어 엔지니어", "결과": "불합격"},
    {"회사명": "회사C", "직무": "제품 관리자", "결과": "진행중"}
]

with open("applications.csv", 'w', newline="") as f:
    fieldnames = ["회사명", "직무", '결과']
    writer = csv.DictWriter(f, fieldnames = fieldnames) # CSV 작성 도구 생성 (딕셔너리)
    writer.writeheader()
    writer.writerows(data) # 여러 행을 한꺼번에 작성

with open("applications.csv", 'r') as f_open:
    reader = csv.DictReader(f_open)
    for row in reader:
        print(row)
````

### **문제 3: `.log` 파일 읽기**

다람쥐는 프로그램 실행 로그를 확인하려고 합니다.

1. `system.log` 파일을 다음 내용으로 생성하세요.
    
    ```
    [INFO] 2024-01-01 12:00:00: 시스템 시작
    [WARN] 2024-01-01 12:05:00: 메모리 사용량 증가
    [ERROR] 2024-01-01 12:10:00: 프로그램 종료
    ```
    
2. 파일을 읽어와 `WARN` 또는 `ERROR`가 포함된 줄만 출력하세요.

````python
with open("./quest/system.log", 'w') as f:
    f.write("[INFO] 2024-01-01 12:00:00: 시스템 시작\n[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n[ERROR] 2024-01-01 12:10:00: 프로그램 종료")

with open("./quest/system.log", 'r') as f_op:

    while True:
        f_line = f_op.readline()
        if '[INFO]' in f_line:
            continue
        
        elif not f_line:
            break

        print(f_line)
````

````python
# 로그파일쓰기

with open('system.log', 'w') as f:
    log_content = [
        "[INFO] 2024-01-01 12:00:00: 시스템 시작\n",
        "[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n",
        "[ERROR] 2024-01-01 12:10:00: 프로그램 종료"
    ]
    f.writelines(log_content)

with open('system.log', 'r') as f:
    for line in f:
        if "WARN" in line or "[ERROR]" in line:
            print(line)
````

2. 수정 작업:

````python
with open("./quest/resume.html", 'w') as f:
    f.write("""
        <html>
            <head><title>다람쥐의 이력서</title></head>
            <body>
                <h1>다람쥐의 이력서</h1>
                <p>경력: 신입</p>
            </body>
        </html>
    """)
````

````python
with open("./quest/resume.html", 'r') as f:
    contents = f.read()

print(contents)
````
