# 다중상속

## 먼저 상속받은 클래스가 우선임

```python
class Daram:
    national = 'korea'

    def greeting(self):
        return 'hello. This is Python'
    
class Runner(Daram):
    
    def __init__(self, subject):
        self.subject = subject

    def run(self):
        return 'I am Running ' + self.subject
    
class Stupid(Runner):
    def __init__(self, name, subject):
        Runner.__init__(self, subject)
        self.name = name

    def greeting(self):
        return '안녕 내이름은 ' + self.name
    
stupid = Stupid('다람', 'Everyday')
print(stupid.greeting())
print(stupid.run())
```

```python
#  부모클래스 선언
class Person:
    def __init__(self, name, phone=None): # phone의 기본값은 None이므로 인자를 안넣어도 됨
        self.name = name
        self.phone = phone

    # 인스턴스 출력시 실행되는 메서드
    def __str_(self):
        return f'Person {self.name}, {self.phone}'
    
# 자식클래스 선언
class Employee(Person):
    def __init__(self, name, phone, position, salary):

        super().__init__(name, phone)  # == Person.__init__(self, name, phone)
        self.position = position
        self.salary = salary

p1 = Person('다람지', '010-1234-5678')
print(p1.name)
print(p1)

m1 = Employee('다람쥐', '010-2345-6789', '직원', '1_0_000_000')
m2 = Employee('쥐람다', '010-5678-8976', '직원', '5_0_000_000')

# m1의 이름과 직책 출력
print(m1.name, m1.position)
# m1의 인스턴스 호출
# print(m1.name, m1.posiiton)
print(m1)
print(m2)
```
