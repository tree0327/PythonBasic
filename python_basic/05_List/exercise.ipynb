```python
location = [ '서울', '부산', '인천', '광주', '울산' ]
print(location[-2:]) # 광주, 울산
print(location[-4:]) # 부산 인천 광주 울산
print(location[1::2]) # 부산 광주
print(location[2::-1]) # 인천 부산 서울
```

```python
nation = ( '대한민국', '미국', '중국', '일본', '프랑스' )

print(nation[1:]) # 미국 중국 일본 프랑스
print(nation[-3:]) # 중국 일본 프랑스
print(nation[1:5:2]) # 미국 일본
print(nation[::-1]) # 프랑스 일본 중국 미국 대한민국
```
