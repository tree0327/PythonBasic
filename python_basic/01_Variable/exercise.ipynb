### 특정 값을 무시

```python
x, _, z = (1, 2, 3)
print(x)    # 1
print(z)    # 3
```


### 앞에 *을 추가로 붙여 여러 개의 값 무시


```python
x, *_, y = (1, 2, 3, 4, 5)
print(x)    # 1
print(y)    # 5
```
