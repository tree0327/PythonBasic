## 파이썬은 변수를 LEGB 순서로 찾는다

```python
# namespace.py로 가정

# 전역 변수
variable = 'global variable'
print('전역 ', variable)

def outer_function():
    # 외부 변수
    variable = "outer variable"
    print("지역(외부 함수) ", variable)

    def innter_function():
        variable = "innter variable"
        print("지역(내부 함수) ", variable )

    innter_function()

outer_function()
```

```python
class TestClass:
    # 클래스 네임스페이스
    variable = 'class variable'

    def __init__(self, value):
        self.variable = value

    def class_function(self):
        variable = 'local variable'
        print('클래스 지역: ', variable)

obj = TestClass('instance variable')
print('인스턴스: ', obj.variable)
obj.class_function()
```

```python
g_variable = 'global variable'

def modify_global():
    global g_variable
    g_variable = 'global modified in function'

print(g_variable)
modify_global()
print(g_variable)
```

```python
class Character:
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.__type = type

    def get_type(self):
        return self.__type

    def set_type(self, type):
        character_types = ['전사', '법사', '치유사']
        for given_type in character_types:
            if(type == given_type):
                self.__type = type
                return

            raise ValueError("잘못된 캐릭터 타입을 선택하였습니다.")

my_character = Character('카피바라', '법사')

print(my_character.nickname)
# print(my_character.__type) 바로 접근 안됨
print(my_character.get_type())
my_character.set_type('법사')
my_character.get_type()

# my_character2 = Character('capybara', '힐러')
# print(my_character2.nickname)
# print(my_character2.get_type())
```
