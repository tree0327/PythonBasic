## 파이썬은 변수를 LEGB 순서로 찾는다

```python
# namespace.py로 가정

# 전역 변수
variable = 'global variable'
print('전역 ', variable)

def outer_function():
    # 외부 변수
    variable = "outer variable"
    print("지역(외부 함수) ", variable)

    def innter_function():
        variable = "innter variable"
        print("지역(내부 함수) ", variable )

    innter_function()

outer_function()
```

```python
class TestClass:
    # 클래스 네임스페이스
    variable = 'class variable'

    def __init__(self, value):
        self.variable = value

    def class_function(self):
        variable = 'local variable'
        print('클래스 지역: ', variable)

obj = TestClass('instance variable')
print('인스턴스: ', obj.variable)
obj.class_function()
```

```python
g_variable = 'global variable'

def modify_global():
    global g_variable
    g_variable = 'global modified in function'

print(g_variable)
modify_global()
print(g_variable)
```

```python
# private : 숨겨서 바로 접근하지 못하게 함
class Character:

    def __init__(self, nickname, type):
        self.nickname = nickname # 들어오는 닉네임을 자기 자신의 닉네임에 저장
        self.type = type # 들어오는 타입을 숨겨서 바로 접근하지 못하게 함

    # 숨겨진 변수의 값을 읽을 때 사용
    def get_type(self):
        return self.type

# 숨겨진 변수의 값을 설정할 때 사용
    def set_type(self, type):
        character_types = ['전사', '법사', '치유사']
        for given_type in character_types:
            if (type == given_type):
                self.type = type
                return

        raise ValueError('잘못된 캐릭터 타입을 선책하셨습니다.')

my_character = Character("카피바라", '전사') # 매개변수는 init 매서드의 매개변수 갯수만큼

print(my_character.nickname)
print(my_character.type) # 에러남 숨겨두어서 찾을 수 없음
print(my_character.type) # 에러남 숨겨두어서 찾을 수 없음
print(my_character.get_type())
my_character.set_type('법사')
print(my_character.get_type())
#my_character.set_type('흑마법사') # 에러남 : 허용된 type이 아니기 때문에 
```
